```
import  requests
webdata=requests.get('https://www.php.net/')
#  this will send  request to server  store response
print(webdata)
```

```
#  now we need html data only
myhtmldata=webdata.content
#print(myhtmldata)
#  Now you can apply  BeautifulSoup  
from  bs4   import  BeautifulSoup
soup=BeautifulSoup(myhtmldata,'html5lib')
#  souping  
textdata=soup.get_text()
newdata=textdata.split()
#  string  to  list 
#print(newdata)
import nltk
nltk.download('stopwords')
#  now listing  all stopwords 
from  nltk.corpus  import  stopwords
stw=stopwords.words('english')
print(stw)


```

```
removedata=[i for   i  in newdata if  i.lower() not in stw]
#removedata
```

```
words=[]
c=[]
for  i  in  removedata:
  if i in words:
    continue
  else :
    ct=removedata.count(i)
    words.append(i)
    c.append(ct)
print(words)
print(c)
import matplotlib.pyplot as plt
plt.xlabel('words')
plt.ylabel('counts')
#plt.bar(words,c)
plt.pie(c,labels=words,autopct='%1.1f%%')
plt.show()
```
